```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define simulation function

def simulate_(lambda_, n0, T):
    
    return [lambda_**t * n0 for t in range(T)]

```

```python
# Plot geometric dynamics for various values of lambda > 1

T = 10
n0 = 1

plt.figure(figsize = (10,8))
plt.plot(range(T), simulate_(0.95, n0, T), marker = 'X', label=r'$\lambda = 0.95$', c= 'b')
plt.plot(range(T), simulate_(1.2, n0, T), marker = 'X', label = r'$\lambda = 1.2$', c = 'red')
plt.plot(range(T), simulate_(1.5, n0, T), marker = 'X', label = r'$\lambda = 1.5$', c= 'green')
plt.legend()
plt.xlabel(r'$T$', fontsize=10)
plt.ylabel('Population', fontsize=10)
plt.yscale("log")
plt.savefig('log_dynamics');
```
